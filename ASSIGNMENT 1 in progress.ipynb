__b)__ Define a function the find the __angle__ between these vectors rounded to __3 decimal places__

```python
#find angle between vectors using dot product

def ang(u,v,dp):
    return np.round(np.degrees(np.arccos(u.dot(v)/norm(u,dp)/norm(v,dp))),dp)
    
```

```python
# calculate angle to 3 decimal places
ang(u,v,3)
```

# Question 3

Given the 3D-vectors
$$
\vec{p} = 4.5\hat{\imath} + 1.6\hat{\jmath} + 2.8\hat{k} \qquad 
\vec{q} = 1.4\hat{\imath} + 4.6\hat{\jmath} - 7.1\hat{k}
$$
answer the following:


__a)__ Define a function __Z(x,y,p,q,B)__ which describes the surface spanned by the vectors $\vec{p}$ and $\vec{q}$ containing the __base-point__ $B(a,b,c)$

```python
def v3d(a,b,c,dp): # use dp to choose number of decimal places to round to.
    return np.round(np.array([a,b,c]),dp)
```

```python
def Z(x,y,p,q,B): # The function requires 2 vectors u & v and a base-point B and x,y-coordinates
    n1,n2,n3=np.cross(p,q)
    a,b,c = B
    z=c-n1/n3*(x-a)-n2/n3*(y-b)
    return np.round(z,3) # round to 3 d.p.
```

__b)__ If $B(-4.5,2.3,1.5)$ is a point in the plane spanned by $\vec{p}$ and $\vec{q}$, create a __3D plot__ of the surface




__NOTE:__ The plot should also display the vectors $\vec{p}$, $\vec{q}$ and $\vec{p}\times\vec{q}$ along with the points $B$ and $B+2\vec{p}-\vec{q}$

```python
p = v3d(4.5, 1.6, 2.8, 3)
q = v3d(1.4, 4.6, -7.1, 3)
n=np.cross(p,q)
B = v3d(-4.5, 2.3, 1.5, 3)
P = B + 2*p + 1*q
p,q,n,B,P
```

```python
# create axes

a,b,c=B
px,py,pz=P
```

```python
X = np.linspace(a-20, a+20, 101) # Since a = -4.5, the x-axis goes from -4.5 - 20 to -4.5 + 20 with 101 points
Y = np.linspace(b-20, b+20, 101) # Since b = 2.3, the y-axis goes from 2.3 - 20 to 2.3 + 20 with 101 points

x,y = np.meshgrid(X,Y)  # 101 x 101 = 10201 points in x,y plane
```

```python
@interact(el=(0,360,5) , az = (0,360,5)) # setup for moving plot

def MovingPlot1(el=20, az=110):  # start points for moving plot

  fig = plt.figure(figsize=(12,9)) # Create a figure environment we call fig, make it 12" x 9" in size
  ax = fig.gca(projection='3d') # Create 3D axes in this figure and call these axes ax

  # The surface is plotted using the function below
  ax.plot_surface(x, y, Z(x,y,p,q,B), color='c', alpha =0.5)

  # The command for plotting 3D vectors is similar, we just use ax.quiver() instead of plt.quiver()
  ax.quiver(a, b, c, p[0], p[1], p[2], color='r', label='$\\vec{p}$')
  ax.quiver(a, b, c, q[0], q[1], q[2], color='g', label='$\\vec{q}$')
  ax.quiver(a, b, c, n[0], n[1], n[2], color='b', label='$\\vec{n}$')

  # The points are plotted using ax.plot() as follows N.B. note the square brackets [] !!
  ax.plot([a], [b], [c], 's', color='k', label='$B$')
  ax.plot([px], [py], [pz], 'D', color = 'y', label='$P$' )

  # Setting the axis-labels is slightly different, we must use ax.set_xlabel() etc.
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_zlabel('z')

  # Setting the axis-limits is also slightly different, we must use ax.set_xlim() etc.
  ax.set_xlim(a-10, a+10)
  ax.set_ylim(b-10, b+10)
    
  ax.view_init(elev=el, azim=az)

  ax.legend()
```
